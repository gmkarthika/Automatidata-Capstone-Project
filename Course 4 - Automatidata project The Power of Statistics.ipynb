```python
# descriptive stats code for EDA
taxi_data.describe(include='all')
```

We are interested in the relationship between payment type and the total fare amount the customer pays. One approach is to look at the average total fare amount for each payment type.

```python
taxi_data.groupby('payment_type')['total_amount'].mean()
```

Based on the averages shown, it appears that customers who pay in credit card tend to pay a larger total fare amount than customers who pay in cash. However, this difference might arise from random sampling, rather than being a true difference in total fare amount. To assess whether the difference is statistically significant, conduct a hypothesis test.

**Hypothesis testing**
Before conducting hypothesis test, consider the following questions where applicable to complete code response:

Recall the difference between the null hypothesis and the alternative hypotheses. What are the hypotheses for this data project?
**Answer: Null hypothesis** There is a no difference in average total fare between customers who pay using credit cards and customers who pay cash **Alternative hypothesis:** There is a difference in average total fare between customers who pay using credit cards and customers who pay cash

Goal in this step is to conduct a **two-sample t-test.** Below are the steps for conducting a hypothesis test:

1. State the null hypothesis and the alternative hypothesis
2. Choose a signficance level
3. Find the p-value
4. Reject or fail to reject the null hypothesis

Note: For the purpose of this exercise, hypothesis test is the main component of A/B test.

**$H_0$:** There is no difference in the average total fare amount between customers who use credit cards and customers who use cash.

**$H_1$:** : There is a difference in the average total fare amount between customers who use credit cards and customers who use cash.

Choose 5% as the significance level and proceed with a two-sample t-test.

```python
#Get the payment amount for credit card transaction
credit_card = taxi_data[taxi_data['payment_type'] == 1]['total_amount']
#Get the payment amount for cash transaction
cash = taxi_data[taxi_data['payment_type'] == 2]['total_amount']\

#hypothesis test, A/B test
stats.ttest_ind(a = credit_card, b = cash, equal_var=False )
```

Since the p-value is extremely small (much smaller than the significance level of 5%), you reject the null hypothesis. You conclude that there is a statistically significant difference in the average total fare amount between customers who use credit cards and customers who use cash.
